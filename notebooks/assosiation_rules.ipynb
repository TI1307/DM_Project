## **Association Rules**


```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
Borrowings_Table=pd.read_excel("../data/cleaned_borrowings.xlsx")
```

```python
Borrowings_Table.head()
```

```python
Borrowings_Table.shape
```

```python
Borrowings_transactions = Borrowings_Table.groupby("N° lecteur")["Titre_clean"].apply(list).reset_index()
```

```python
Borrowings_transactions.head()
```

```python
print("Number of transactions:", len(Borrowings_transactions))
print("Number of distinct customers:", len(list(set(Borrowings_transactions["N° lecteur"]))))
```

```python
transactions = Borrowings_transactions["Titre_clean"].tolist()
transactions[:5]
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
borrowing_df = pd.DataFrame(te_ary, columns=te.columns_)
borrowing_df
```

```python
frequent_itemsets_fpg=fpgrowth(borrowing_df, min_support=0.01)
frequent_itemsets_fpg
```

```python
frequent_itemsets_apr=apriori(borrowing_df, min_support=0.01)
frequent_itemsets_apr
```

### Generate Rules


```python
# Make sure you use use_colnames=True to get actual book titles
frequent_itemsets_fpg = fpgrowth(borrowing_df, min_support=0.01, use_colnames=True)
frequent_itemsets_apr = apriori(borrowing_df, min_support=0.01, use_colnames=True)

rules_fpg=association_rules(frequent_itemsets_fpg, metric="confidence", min_threshold=0.01)
rules_apr=association_rules(frequent_itemsets_apr, metric="confidence", min_threshold=0.01)
rules_fpg
```

```python


# ============================================================================
# STEP 2: Display rules with actual book titles
# ============================================================================

def show_rules(rules_df):
    """Display association rules in readable format with book titles"""
    print(f"\nTotal Rules: {len(rules_df)}")
    print("="*100)
    
    for idx, row in rules_df.iterrows():
        # Get the actual book titles from antecedents and consequents
        antecedents_books = list(row['antecedents'])
        consequents_books = list(row['consequents'])
        
        print(f"\nRule {idx+1}:")
        print(f"  IF a reader borrows:")
        for book in antecedents_books:
            print(f"    - {book}")
        print(f"  THEN they will also borrow:")
        for book in consequents_books:
            print(f"    - {book}")
        print(f"\n  Metrics:")
        print(f"    Support: {row['support']:.2%} | Confidence: {row['confidence']:.2%} | Lift: {row['lift']:.2f}")
        print("-"*100)

# Use it:
show_rules(rules_apr)


```
